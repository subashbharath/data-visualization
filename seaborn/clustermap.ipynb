```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flight=sns.load_dataset("flights")
flight
```

```python
flight1=flight.pivot("month","year","passengers",)
flight1
```

```python
#both row and column cluster
'''parameters
row_cluster=false onnlyrow cluster
col_cluster-false only cluster
cmap= color same as palette
annot =values of corr
line width=0.1 to 1 width of clustermap
figsize=size of the figure
standard_scale=1 #standarize the value 1 for column and 0 for row
z_score=zscore values based on rows and column (1 for column and 0 for row)'''
sns.clustermap(flight1)
```

```python
#with annot
sns.clustermap(flight1,annot=True,fmt="d")
```

```python
#only row cluster
sns.clustermap(flight1,col_cluster=False)
```

```python
#only column cluster
sns.clustermap(flight1,row_cluster=False)
```

```python
#cmap same as palette
sns.clustermap(flight1,cmap="spring")
```

```python
#line width with figure size
sns.clustermap(flight1,linewidth=0.5,figsize=(10,6))
```

```python
#standarize the value
sns.clustermap(flight1,standard_scale=1)
```

```python
#1 for column and 0 for row
sns.clustermap(flight1,z_score=1,cmap="Accent")
```
