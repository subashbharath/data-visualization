```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#simple distplot
'''parametres
color= color of the kde and histrograms
vertical=True-horizontalbar
kde=false-only histofram
hist=false-only kde
rug=true
bins=no.of bars in histogram or range interval'''
num=np.random.randn(150)
sns.distplot(num,color="red")
plt.title("distplot")
plt.show()
```

```python
#with x axis
label_x=pd.Series(num,name="varaible_x")
sns.distplot(label_x)
plt.title("distplot")
plt.show()
```

```python
#horizontal bar
sns.distplot(label_x,vertical=True,color="red")
plt.title("distplot")
plt.show()
```

```python
#only kde
sns.distplot(label_x,hist=False,color="red")
plt.title("distplot")
plt.show()
```

```python
#only kde with rug
sns.distplot(label_x,hist=False,rug=True,color="red")
plt.title("distplot")
plt.show()
```

```python
#only histogram with 20 bars
sns.distplot(label_x,kde=False,color="red",bins=20)
plt.title("distplot")
plt.show()
```
