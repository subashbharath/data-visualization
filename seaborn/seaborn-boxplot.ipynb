```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset("tips")
tips
```

```python
#simple box plot with one axis (xaxis)
'''parameters
x= xaxis colunmname
y= yaxis colunm name
data=dataset object
hue=interpolartion
palatte=colors of each image
color=all in same color
order= order of xaxis
linewidth=width of line
orient=h_ horitonal'''
sns.boxplot(x=tips["total_bill"])
plt.title("boxplot")
plt.show()
```

```python
#simple boxplot with two axis
sns.boxplot(x="sex",y="total_bill",data=tips)
plt.title("boxplot")
plt.show()
```

```python
#with hue
sns.boxplot(x="sex",y="total_bill",data=tips,hue="smoker")
plt.title("boxplot")
plt.show()
```

```python
#with hue and with palette=
sns.boxplot(x="sex",y="total_bill",data=tips,hue="smoker",palette="prism")
plt.title("boxplot")
plt.show()
```

```python
#linewidth
sns.boxplot(x="day", y="total_bill", hue="time",data=tips, linewidth=2.5)
```

```python
#order
sns.boxplot(x="time", y="tip", data=tips,order=["Dinner", "Lunch"])
```

```python
#orientation
iris = sns.load_dataset("iris")
sns.boxplot(data=iris, orient="h", palette="Set2")
```

```python
#orientation
iris = sns.load_dataset("iris")
sns.boxplot(data=iris, orient="v", palette="Set2")
sns.swarmplot(data=iris)
```
