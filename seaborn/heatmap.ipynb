```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flight=sns.load_dataset("flights")
flight
```

```python
flight1=flight.pivot("month","year","passengers",)
flight1
```

```python
#simple heatmap
'''vmin=minium value,
    vmax=max value,
    cmap=colors,
    center=data.iloc["f1","f2"]
    annot=None,
    fmt='d',,
    linewidths=0,
    linecolor='white',
    cbar=True'''

sns.heatmap(flight1)
```

```python
#without cbar 
sns.heatmap(flight1,cbar=False)
```

```python
sns.heatmap(flight1,annot=True,fmt="d")
```

```python
#vmax and min value
sns.heatmap(flight1,vmin=0,vmax=1)
```
