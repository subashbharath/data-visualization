```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
#simple violinplot
'''only x axis'''
sns.violinplot(x=tips["total_bill"])
```

```python
#both xaxis and yaxis
'''x=xaxis colunm name
y=yaxis colunm name
data=dataset object
color= all in same color
hue=interploation
linewidth= points line width
palllete=color
orient=h or v
dodge=it spilit the hue values (disimilar value)
order=order of x axis
scale=scale of gthe plot'''
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
#with hue
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex")
```

```python
#hue with split
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

```python
#with sacle
sns.violinplot(x="day",y="total_bill",data=tips,inner='box',scale='area')
```
