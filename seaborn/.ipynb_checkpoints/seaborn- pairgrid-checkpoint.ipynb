```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris=sns.load_dataset("iris")
```

```python
#simple pairgrid
'''parameters
var=#for only selected no.of features
x_var=#for only selected no.of features (xaxis)
y_var=#for only selected no.of features (yaxis)
map_upper=above diagonal
map_lower=below diagonal

'''
x=sns.PairGrid(iris)
x.map(plt.scatter)
```

```python
#dialong diagram with histogram
'''x.map_diag= diagonal graph
map_offdiag= other than diagonal image'''
x=sns.PairGrid(iris,hue="species")
x.map_diag(plt.hist)
x.map_offdiag(plt.scatter)

#adding legend
x.add_legend()
```

```python
#modifiying the histogram
x=sns.PairGrid(iris,hue="species")
x.map_diag(plt.hist,histtype="step",bins=20,rwidth=0.5)
x.map_offdiag(plt.scatter)

#adding legend
x.add_legend()
```

```python
#for only selected no.of features
x=sns.PairGrid(iris,vars=["petal_width","petal_length"],hue="species")
x.map_diag(plt.hist,edgecolor="black")
x.map_offdiag(plt.scatter,edgecolor="black")
```

```python
#selecting xaxis and y axis
x=sns.PairGrid(iris,x_vars=["petal_width","petal_length"],y_vars=["sepal_width","sepal_length"],hue="species")
x.map(plt.scatter)
```

```python
#differnt graphs above nas below the diagonal
x=sns.PairGrid(iris,hue="species")
x.map_diag(plt.hist)
x.map_upper(plt.scatter)
x.map_lower(sns.kdeplot)
x.add_legend()
```
