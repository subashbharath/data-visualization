```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
#with rows and col
x=sns.FacetGrid(tips,row="time",col="smoker")
#mapping the feature
x.map(plt.hist,"total_bill")
```

```python
#only col and changiung the order of col
x=sns.FacetGrid(tips,col="day",col_order=["Sat","Sun","Thur","Fri"])
#mapping the feature
x.map(plt.hist,"total_bill")
```

```python
#only with row and changing the row order
x=sns.FacetGrid(tips,row="day",row_order=["Sat","Sun","Thur","Fri"])
#mapping the feature
x.map(plt.hist,"total_bill",color="red")
```

```python
#scatter plot with facegrid
#with rows and col
x=sns.FacetGrid(tips,row="time",col="smoker")
#mapping the feature
x.map(plt.scatter,"total_bill","tip")
```

```python
#with regression plot
x=sns.FacetGrid(tips,row="time",col="smoker")
#mapping the feature
x.map(sns.regplot,"total_bill","tip")
```

```python
#with hue #bivaraiate analaysis
x=sns.FacetGrid(tips,hue="smoker")
#mapping the feature
x.map(plt.scatter,"total_bill","tip").add_legend()
```
