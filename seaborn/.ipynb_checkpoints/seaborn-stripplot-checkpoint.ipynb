```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset("tips")
tips
```

```python
#simple stripplot
""""parameters
x=xaxis colunm name
y=yaxis colunm name
data=dataset object
color= all in same color
jiiter=true or sample value 0-1
hue=interploation
linewidth= points line width
palllete=color
orient=h or v
dodge=it spilit the hue values (disimilar value)
order=order of x axis
"""
#with only xaxis
sns.stripplot(x=tips["tip"],color="c")
```

```python
#with only xaxis and yaxis
sns.stripplot(x="total_bill",y="day",data=tips,order=["Sat","Fri","Sun","Thur"])
```

```python
#with jitter
sns.stripplot(x="total_bill",y="day",data=tips,jitter=True)
```

```python
#vertical automated
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
```

```python
#with hue
sns.stripplot(x="total_bill",y="day",data=tips,hue="sex",linewidth=1)
```

```python
sns.stripplot(x="total_bill",y="day",data=tips,hue="sex",linewidth=1,jitter=True,dodge=True)
```
