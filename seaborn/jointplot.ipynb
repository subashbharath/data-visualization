```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
# simple pairplot
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
#with reg plot and color
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg",color="green")
```

```python
#with stats model
from scipy.stats import spearmanr
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg",color="green",stat_func=spearmanr)
```
