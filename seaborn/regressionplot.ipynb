```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
```

```python
#simple regplot
sns.regplot(x="total_bill",y="tip",data=tips,color="c")
```

```python
#scatter plot using regplot
sns.regplot(x="total_bill",y="tip",data=tips,fit_reg=False,marker="+")
```

```python
#modifyiong the scatter and line
sns.regplot(x="total_bill",y="tip",data=tips,scatter_kws={"marker":"*",'color':"blue"},line_kws=('line'='red','linewidth'=2.5))
```
